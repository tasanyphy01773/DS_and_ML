```python
#random_variables
missing_days = 0

#defining incentive
inc_rank_A = 0.04
inc_rank_B = 0.04
inc_rank_C = 0.05
inc_rank_D = 0.1

#definign penalty
pen_rank_A = 0.01
pen_rank_B = 0.01
pen_rank_C = 0.02
pen_rank_D = 0

class employee:
  def __init__(self, name, unique_id, position, experience, rank, basic_salary):
    self.name = name
    self.id = unique_id
    self.position = position
    self.experience = experience 
    self.rank = rank
    self.basic_salary = basic_salary

  def get_name(self):
    return self.name
  
  def get_incentive_and_penalty(self):
    if self.rank == "A":
      self.incentive = self.basic_salary*inc_rank_A
      self.penalty = self.basic_salary*pen_rank_A

    elif self.rank == "B":
      self.incentive = self.basic_salary*inc_rank_B
      self.penalty = self.basic_salary*pen_rank_B

    elif self.rank == "C":
      self.incentive = int(self.basic_salary*inc_rank_C)
      self.penalty = self.basic_salary*pen_rank_C
        
    else:
      self.incentive = self.basic_salary*inc_rank_D
      self.penalty = self.basic_salary*pen_rank_D
    


  def get_id(self):
    return self.id

  def total_salary(self):
    self.get_incentive_and_penalty()
    total_salary = ((self.basic_salary + (self.experience * self.incentive)) - missing_days*self.penalty)
    return total_salary


  #get first name with all lowercase
  def get_first_name(self):
      name = self.name 
      name_list = list(name.split(" "))
      first_name =  name_list[0].lower()
      return first_name
  
  def round_position(self):
    position = self.position
    postion_list = list(position.split(" "))
    new_position = (postion_list[0]+postion_list[1]).lower()
    return new_position



  def get_email(self):
    return f"{self.get_first_name()}.{self.round_position()}@{company_name.lower()}.com"


  def get_reciept(self):
    print(" ")
    print("..................................................... ")
    print("Salary reciept of the employee")
    print("..................................................... ")
    print(f'Name: {self.name}')
    print(f'Id: {self.id}')
    print(f'Position: {self.position}')
    print(f"Email: {self.get_email()}")
    print(f'Experience: {self.experience} years')
    print(f"Missed workday in {month}: {missing_days} ")
    print(f'Total Salary of {month} : {self.total_salary()}')
    print("..................................................... ")

print("..................................................... ")
print("Salary System Management")
print("Team Sigma")
print("..................................................... ")
print(" ")


company_name = input("Name of the company: ")
n = int(input("Number of Employees: "))
employee_info = []
employee_list = []




for i in range(n):
  print(" ")
  print("..................................................... ")
  print(f'Enter data for employee {i+1}')
  print("..................................................... ")
  name = str(input("Full Name: "))
  unique_id = str(input("Id: "))
  position = str(input("Job Position: "))
  experience = int(input("Experience(in year): "))
  rank = input("Rank(A/B/C/D)[Must be capital]: ")
  basic_salary = int(input("Basic Salary: "))
  employee_list.append(name)
  employee_info.append(employee(name, unique_id, position, experience, rank, basic_salary))

while True:
  print(" ")
  print("..................................................... ")
  reciept = input("Do you want reciept(yes/no): ")
  if reciept.lower() == 'yes':
    print("..................................................... ")
    print('Provide information to get reciept.')
    print("..................................................... ")
    full_name = str(input("Enter full name of your employee: "))
    x = employee_list.index(name)
    month = str(input("Month: "))
    missing_days = int(input(f"Missed workdays in {month}: "))
    employee_info[x].get_reciept()
  elif reciept.lower() == 'no':
    break
  else:
    print(f"There is no employee named {full_name}")
```
