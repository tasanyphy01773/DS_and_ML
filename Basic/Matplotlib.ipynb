```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
Month = ['Jan', 'Feb', 'Mar','Apr', 'May', 'Jun', 'Jul', 
         'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

Sany = [4,5,1,3,5,6,2,7,9,1,3,4]
Sadaf = [2,7,4,7,9,1,3,4,1,3,4,8]
```

```python
plt.plot(Month, Sany, 'red', label ='Sany', marker ='o')
plt.plot(Month, Sadaf, 'blue', label='Sadaf', marker ='.')
plt.xlabel("Month")
plt.ylabel("Working Hour")
plt.title("Working hour of employee")
plt.legend()
plt.show()
```

```python

plt.subplot(1,2,1)
plt.plot(Month, Sany, 'red', label ='Sany', marker ='o')
plt.xlabel("Month")
plt.ylabel("Working Hour")
plt.title("Working hour of employee")
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,2)
plt.plot(Month, Sadaf, 'blue', label='Sadaf', marker ='.')
plt.xlabel("Month")
plt.ylabel("Working Hour")
plt.title("Working hour of employee")
plt.margins(2)
plt.legend()
plt.show()
```
